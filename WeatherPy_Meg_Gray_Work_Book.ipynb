Cloudiness on October 11, 2018(EST), there does not appear to be any significant correlation between the percentage of clouds and changes in latitude of cities North to South. If we took elevation of cities into account, there may be a correlation that would be significant compared this scenario of only latitude vs. cloudiness.

* For the City Latitude vs Maximum Temperature scatterplot, there is a correlation to maximum temperature and latitude. The max temperature increases the closer the cities are to the equator. The temperatures also go down on either side of the hemisphere, the further the cities are from the equator.

The scatterplot does appear to indicate that the cities closer the equator experienced lower wind speeds overall when compared to some cities in the Northern and Southern hemisphere. Finally, those cities with the heighest wind speeds overall were cities above 40 degrees latitude, and below -30 degrees latitude.

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create dictionary of all relevant values we are using.

d = {'City': cities, 
     'Latitude': lat, 
     'Longitude': lon, 
     'Temperature': temp, 
     'Clouds':clouds, 
     'Humidity':humidity, 
     'MaxTemp':Max_Temp,
     'WindSpeed':Wind_Speed,
     'Date':date
    }
```

```python
# Create a datafame of data in dictionary. 
Raw_DataFrame = pd.DataFrame.from_dict(d, orient='index')

# Data was not oriented correctly in dataframe, so need to use transpose function to orient the dataframe correctly.
df= Raw_DataFrame.transpose()
df['Date'] = pd.to_datetime(df['Date'],unit='s')
df.head()
```

```python
# Create a single variable to host the timestamp that is needed for the scatterplots.
date_df = df['Date']
Time_stamp = date_df[0]
```

```python
# Drop the cities that do not have values from the dataframe.
Final_Weather_df = df.dropna()
Final_Weather_df.head()
```

```python
# Export dataframe to csv file.
Final_Weather_df.to_csv(path_or_buf='WeatherPyResults.csv', encoding='utf-8', index=False)
```

#### Latitude vs. Temperature Plot

```python
# Create a scatter plot of the cities latitudes versus the maximum temperatures and label all axis and title plot.

plt.scatter(x=df.Latitude,y=df.MaxTemp, edgecolor="black")
plt.legend()
plt.title(f"City Latitude vs. Maximum Temperature ({Time_stamp})")
plt.xlabel("Latitude")
plt.ylabel("Temperature (Celsius)")
plt.savefig('LatTemp.png', bbox_inches='tight')
```

#### Latitude vs. Humidity Plot

```python
# Create a scatter plot of the cities latitudes versus the humidity and label all axis and title plot.

plt.scatter(x=df.Latitude,y=df.Humidity, edgecolor="black")
plt.legend()
plt.title(f"City Latitude vs. Humidity ({Time_stamp})")
plt.xlabel("Latitude")
plt.ylabel("Humidity %")
plt.savefig('LatHum.png', bbox_inches='tight')
```

#### Latitude vs. Cloudiness Plot

```python
# Create a scatter plot of the cities latitudes versus cloudiness and label all axis and title plot.

plt.scatter(x=df.Latitude,y=df.Clouds, edgecolor="black")
plt.legend()
plt.title(f"City Latitude vs. Cloudiness ({Time_stamp})")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness %")
plt.savefig('LatCloud.png', bbox_inches='tight')
```

#### Latitude vs. Wind Speed Plot

```python
# Create a scatter plot of the cities latitudes versus the wind speed and label all axis and title plot.

plt.scatter(x=df.Latitude,y=df.WindSpeed, edgecolor="black")
plt.legend()
plt.title(f"City Latitude vs. Maximum Wind Speed ({Time_stamp})")
plt.xlabel("Latitude")
plt.ylabel("Maximum Wind Speed (mph)")
plt.savefig('LatWind.png', bbox_inches='tight')
```
